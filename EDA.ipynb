```python
# 1 Import Data

import pandas as pd

books = pd.read_csv(r"C:\Users\Admin\Desktop\20260303-DE5M5\data\03_Library Systembook.csv")
customers = pd.read_csv(r"C:\Users\Admin\Desktop\20260303-DE5M5\data\03_Library SystemCustomers.csv")
```

```python
# 2 Remove quotes from Book checkout column
from newscript import clean_column
books = clean_column(books)
```

```python
# 3 Delete null values from Books ID and Customer ID columns
books = books.dropna(subset=['Id'])
books = books.dropna(subset=['Customer ID'])
```

```python
# 4 Change Book ID columns to INT
books['Id'] = books['Id'].astype(int)
books['Customer ID'] = books['Customer ID'].astype(int)
```

```python
# 5 Update 2 weeks to 14 days in Books
books['Days allowed to borrow'] = 14
```

```python
# 6 Change Customers ID column to INT
customers['Customer ID'] = customers['Customer ID'].astype(int)
```

```python
# 7 Delete null values from Customer ID in customers
customers = customers.dropna(subset=['Customer ID'])
```

```python
# 8 Update any incorrect date values
books['Book checkout'] = pd.to_datetime(books['Book checkout'],dayfirst=True,errors='coerce')
books['Book Returned'] = pd.to_datetime(books['Book Returned'],dayfirst=True,errors='coerce')
```

```python
# 9 Update text columns to date
books['Book checkout'] = pd.to_datetime(books['Book checkout'])
books['Book Returned'] = pd.to_datetime(books['Book Returned'])
```

```python
books['Total Days'] = (books['Book Returned'] - books['Book checkout']).dt.days
```

```python
# 10 Export clean data

books.to_csv('Books.csv',index=False)
customers.to_csv('Customers.csv',index=False)
```
